```python
import requests  # used to make HTTP requests to web servers
from bs4 import BeautifulSoup #  for parsing HTML and XML documents
import pickle # for serializing and deserializing Python objects and save Python objects to a file and then load them back into memory later
def scrape_text_from_url(url):
    try:
        # Fetch the HTML content of the URL
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for 4xx or 5xx status codes
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find and extract the text content
        text_content = soup.get_text(separator='\n', strip=True)
        
        return text_content
    except Exception as e:
        print(f"Error occurred while scraping {url}: {e}")
        return None
        
urls = ['https://science.thewire.in/the-sciences/opinion-time-to-rethink-age-restriction-jrf/',
        'https://science.thewire.in/education/dismayed-academics-scientists-iisc-uapa-talk/',
        'https://science.thewire.in/education/opinion-covid-19-education-disruption-recovery/',
        'https://science.thewire.in/education/research-fellows-problems-solutions/',
        'https://science.thewire.in/education/minority-report-caste-gender-representation-iiser-pune/',
        'https://science.thewire.in/the-sciences/review-iisf-science/',
        'https://science.thewire.in/the-sciences/china-high-quality-science-research/',
        'https://science.thewire.in/education/nep-agricultural-university-transform/',
        'https://science.thewire.in/education/sanksrit-language-machine-panini-grammar-rishi-rajpopat/',
        'https://science.thewire.in/health/mp-mbbs-hindi-syllabus-english/',
        'https://science.thewire.in/health/student-suicides-grave-crisis-india-education/',
        'https://science.thewire.in/education/post-pandemic-schooling-rushed-classrooms-terrible-idea/',
        'https://science.thewire.in/the-sciences/gendered-segregated-hostels-science-institutes/',
        'https://science.thewire.in/the-sciences/ai-ml-students-plagiarism/',
        'https://science.thewire.in/the-sciences/interview-the-threat-of-self-censorship-in-science/',
        'https://science.thewire.in/environment/sahara-desert-green-savannah-new-research-explains-why/',
        'https://science.thewire.in/the-sciences/us-uk-gene-editing-therapy-sickle-cell-disease/',
        'https://science.thewire.in/the-sciences/chatgpt-productive-role-science-education/',
        'https://science.thewire.in/environment/wild-varieties-crops-resilient-agriculture-climate-change/',
        'https://science.thewire.in/the-sciences/neanderthals-more-knowable-now-than-ever/',
        'https://science.thewire.in/health/no-compelling-evidence-air-purifiers-respiratory-infections/',
        'https://science.thewire.in/the-sciences/fossils-are-shaped-by-people-does-that-matter/',
        'https://science.thewire.in/the-sciences/water-corridors-helped-homo-sapiens-disperse-out-of-africa/',
        'https://science.thewire.in/the-sciences/pursuing-fusion-power/',
        'https://science.thewire.in/aerospace/interview-nasa-push-track-unexplained-objects-david-spergel/',
        'https://science.thewire.in/the-sciences/the-suns-activity-cycle-is-reaching-its-peak-early-what-it-means/',
        'https://science.thewire.in/the-sciences/discovery-500000-year-old-wooden-structure-ancient-relatives/',
        'https://science.thewire.in/the-sciences/gaia-space-observatory-half-a-million-new-stars-milky-way/',
        'https://science.thewire.in/the-sciences/what-is-an-attosecond-the-time-scale-behind-this-years-physics-nobel-win/',
        'https://science.thewire.in/the-sciences/pierre-agostini-ferenc-krausz-anne-lhuillier-physics-nobel-2023-electrons-study/',
        'https://science.thewire.in/health/why-less-sleep-cannot-make-workers-more-productive/',
        'https://science.thewire.in/health/why-we-shouldnt-lose-our-minds-over-sleep/',
        'https://science.thewire.in/health/breast-cancer-can-bring-financial-ruin-in-india/',
        'https://science.thewire.in/health/are-zombie-cancer-cells-why-the-disease-sometimes-returns/',
        'https://science.thewire.in/health/climate-change-is-driving-up-spread-of-diseases-but-fewer-people-confident-in-vaccines/',
        'https://science.thewire.in/health/ultra-processed-foods-heres-what-the-evidence-actually-says-about-them/',
        'https://science.thewire.in/health/in-hong-kong-researchers-make-progress-in-early-diagnostic-tools/',
        'https://science.thewire.in/health/india-public-health-system-rural-women-domestic-abuse/',
        'https://science.thewire.in/health/long-covid-brain-fog-acquired-form-adhd/',
        'https://science.thewire.in/health/nanoplastics-parkinsons-dementia-new-study/',
        'https://science.thewire.in/health/lack-of-robust-diverse-data-a-hurdle-in-managing-brain-strokes-in-india/']

name = ['file' + str(i) for i in range(1, 42)]
```

```python
# We are going to change this to key: comedian, value: string format
def combine_text(list_of_text):
    '''Takes a list of text and combines them into one large chunk of text.'''
    combined_text = ''.join(list_of_text)
    return combined_text
```

### ADDITIONAL FUNCTIONS FOR DATA CLEANING

```python
def remove_header_footer(text):
    start_index = text.find("Read Next")
    end_index = text.find("Share")
    
    if start_index != -1 and end_index != -1:
        return text[start_index:end_index].strip()
    else:
        return text  # Return original text if header or footer not found
```

```python
round = lambda x: remove_header_footer(x)
data_df = pd.DataFrame(data_df.transcript.apply(round))
data_df
```

```python
def remover(text):
    start_index = text.find("Read Next\n")
    end_index = text.rfind("\n", 0, text.rfind("\n"))  # Find the index of the second last line break
    
    if start_index != -1 and end_index != -1:
        return text[start_index+len("Read Next\n"):end_index].strip()
    else:
        return text  # Return original text if header or footer not found
```

```python
round = lambda x: remover(x)
data_df = pd.DataFrame(data_df.transcript.apply(round))
data_df
```

```python
import emoji

def remove_emojis(text):
    return emoji.demojize(text)
```

```python
round = lambda x: remove_emojis(x)
data_df = pd.DataFrame(data_df.transcript.apply(round))
data_df
```

```python
# Apply a first round of text cleaning techniques
import re #  provides support for regular expressions and powerful tools for pattern matching and string manipulation
import string # functions for formatting strings, manipulating characters

def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# Let's take a look at the updated text
data_clean = pd.DataFrame(data_df.transcript.apply(round1))
data_clean
```

```python
# Apply a second round of cleaning
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
# Let's take a look at the updated text
data_clean = pd.DataFrame(data_clean.transcript.apply(round2))
data_clean
```

### **CountVectorizer's parameters**   
**ngram_range**: This parameter determines the range of n-grams to consider. An n-gram is a contiguous sequence of n items from a given sample of text. For example:  
ngram_range=(1, 1): Only unigrams (single words) will be considered.
ngram_range=(1, 2): Both unigrams and bigrams (pairs of consecutive words) will be considered.
ngram_range=(2, 2): Only bigrams will be considered.  
**min_df**: This parameter specifies the minimum frequency threshold for a term to be included in the vocabulary. If a term appears in fewer documents than min_df, it will be ignored. This can be an absolute count (e.g., min_df=2) or a proportion of the total number of documents (e.g., min_df=0.1 for 10% of documents)  
**max_df**: This parameter specifies the maximum frequency threshold for a term to be included in the vocabulary. If a term appears in more documents than max_df, it will be ignored. This can also be an absolute count or a proportion.

```python
cv = CountVectorizer(stop_words='english', ngram_range=(1, 2), min_df=2, max_df=0.9)
data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names_out())
data_dtm.index = data_clean.index
data_dtm
```

```python
cv = CountVectorizer(stop_words='english', ngram_range=(2, 2), min_df=3, max_df=0.9)
data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names_out())
data_dtm.index = data_clean.index
data_dtm
```
